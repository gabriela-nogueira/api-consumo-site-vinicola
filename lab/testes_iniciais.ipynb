```python
def get_dict_comercializacao():
    url = "http://vitibrasil.cnpuv.embrapa.br/download/Comercio.csv"
    response = requests.get(url)

    csv_data = []
    
    with StringIO(response.text) as csv_file:
        reader = csv.DictReader(csv_file, delimiter=';')
        for row in reader:
            list_of_data = list(dict(row).values())
            data_formated = {}
            for key, value in dict(row).items():
                if list_of_data.index(value) not in [0,1]:  # Check if the key is a digit (Unix timestamp)
                    year = datetime.utcfromtimestamp(int(value)).year
                    data_formated[year] = key
                else:
                    data_formated[value] = key
                    
            csv_data.append(data_formated)

    return csv_data
```

```python
def transform_table_to_dict(table_string):
    # Parse the HTML using BeautifulSoup
    soup = bs4.BeautifulSoup(str(table_string), 'html.parser')

    # Find all rows in the table
    rows = soup.find_all('tr')

    # Initialize an empty list to store dictionaries
    data = []

    # Extract the headers
    headers = [header.get_text() for header in rows[0].find_all('th')]

    # Extract the data from each row and create a dictionary
    for row in rows[1:]:
        values = [value.get_text() for value in row.find_all('td')]
        row_data = dict(zip(headers, values))
        data.append(row_data)

    return data
```
